```python
def clean_table(x):
  if len(x)==114:
    return x[10:112]
  else:
    return 'not match'
    
```

```python
def prepare_table(pre_table):
    # ทำความสะอาดคอลัมน์ใน pre_table
    pre_table['Male'] = pre_table['Male'].apply(lambda x: clean_table(x))
    pre_table['Female'] = pre_table['Female'].apply(lambda x: clean_table(x))
    pre_table['Male_key'] = pre_table['Male_key'].apply(lambda x: clean_table(x))

    # สร้าง table dictionary เพื่อเก็บข้อมูลที่แปลงแล้ว
    table = {'Province': [], 'Province_code': [], 'District': [], 'District_code': [], 'Male': [], 'Female': [], 'Age': []}
    
    for I in range(len(pre_table)):
        table['Province'].extend([pre_table['Province'][I]] * 102)
        table['Province_code'].extend([pre_table['Province_code'][I]] * 102)
        table['District'].extend([pre_table['District'][I]] * 102)
        table['District_code'].extend([pre_table['District_code'][I]] * 102)
        table['Male'].extend(list(pre_table['Male'][I]))
        table['Female'].extend(list(pre_table['Female'][I]))
        table['Age'].extend(list(pre_table['Male_key'][I]))

    # แปลง table dictionary เป็น DataFrame
    table_df = pd.DataFrame(table)
    
    return table_df
```

```python
pre_table_popst_0_prepare = prepare_table(pre_table_popst_0)
```

```python
pre_table_popst_0_prepare['Condition'] = 'แยกตามจำนวนประชากรที่มีชื่ออยู่ในทะเบียนบ้าน'
```

```python
pre_table_popst_0_prepare.to_csv('แยกตามจำนวนประชากรที่มีชื่ออยู่ในทะเบียนบ้าน.csv', index=False)
```
